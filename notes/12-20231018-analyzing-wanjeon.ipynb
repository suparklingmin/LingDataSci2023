```python
import pandas as pd
```

```python
utterances = pd.read_csv('../NIKL_OM_utterances.csv.tar.gz', compression='gzip', on_bad_lines='skip')
del utterances['../NIKL_OM_utterances.csv']
utterances.dropna(inplace=True)
utterances = utterances[['id', 'form', 'speaker_id']]
utterances
```

```python
speakers = pd.read_csv('../NIKL_OM_speakers.csv.tar.gz', compression='gzip', on_bad_lines='skip')
speakers
```

```python
del speakers['../NIKL_OM_speakers.csv']
speakers.dropna(inplace=True)
speakers = speakers[['speaker_id', 'age', 'sex']]
speakers
```

```python
# 발화 정보에 발화자 정보 통합하기
# "speaker_id"라는 공통 열을 통해 두 데이터프레임을 합치기
df = utterances.merge(speakers, on="speaker_id")
df
```

```python
df.to_csv('../data/NIKL_OM_form_age_sex.csv', index=False)
```

```python
def count_wanjeon(df):
  output = {}
  output['완전'] = df['form'].str.count(r'\b완전\b').sum()
  output['아주'] = df['form'].str.count(r'\b아주\b').sum()
  return pd.Series(output)
```

```python
counts_overall = count_wanjeon(df)
counts_overall
```

```python
counts_overall.div(counts_overall.sum()) * 100
```

```python
def get_percent(series):
    return series.div(series.sum(axis=1), axis=0) * 100
```

```python
counts_by_age = df.groupby('age').apply(count_wanjeon)
counts_by_age
```

```python
get_percent(counts_by_age)
```

```python
counts_by_sex = df.groupby('sex').apply(count_wanjeon)
counts_by_sex
```

```python
get_percent(counts_by_sex)
```

```python
counts_by_age_sex = df.groupby(['age', 'sex']).apply(count_wanjeon)
counts_by_age_sex
```

```python
get_percent(counts_by_age_sex)
```
