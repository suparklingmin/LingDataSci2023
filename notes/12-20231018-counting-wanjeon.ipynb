```python
# 데이터프레임 처리
import pandas as pd
```

```python
# pd.read_csv() csv 파일을 데이터프레임으로 읽기
utterances = pd.read_csv('../NIKL_OM_utterances.csv.tar.gz', compression='gzip', on_bad_lines='skip', encoding_errors='ignore')
utterances
```

```python
# 불필요한 column 지우기
del utterances['../NIKL_OM_utterances.csv']

# 값이 없는 row/column 지우기
utterances.dropna(inplace=True)

# id, form, speaker_id column만 남기고 나머지는 지우기
utterances = utterances[['id', 'form', 'speaker_id']]
utterances
```

# 오늘의 목표

1. 코퍼스에서 '완전', '완전히', '아주' 세 단어의 사용 빈도를 알아본다.
    + '완전연결' - 명사구 vs. '완전 좋음' - 부사

```python
# 완전/완전히/아주 세 단어 중 하나가 포함된 발화 찾기
forms = utterances['form'][utterances['form'].str.contains(r'\b(?:완전히|완전|아주)\b', regex=True)]
forms
```

fit, fat을 모두 찾고자 할 때

r'f(i|a)t'

```python
import re

print(re.findall(r'f(i|a)t', 'fit')) # grouping o

print(re.findall(r'f(?:i|a)t', 'fit')) # grouping x
```

+ '완전 ㅇㅇ' 꼴의 문형에서 ㅇㅇ에 해당하는 것을 가져오기

```python
from collections import Counter
print(Counter(forms.str.findall(r'\b완전 (\w+)').sum()))
```

```python
from collections import Counter

keywords = ('완전', '완전히', '아주')

# dict 초기화
collocations_dict = {} 

# '완전 ㅇㅇ' 꼴에서 'ㅇㅇ'에 해당하는 부분을 모아서 출현 횟수를 세기
for word in keywords:
    collocations_dict[word] = pd.Series(Counter(forms.str.findall(fr'\b{word} (\w+)').sum()))

collocations = pd.DataFrame(collocations_dict)
collocations
```

```python
# NaN 값을 0으로 바꾸기
collocations[collocations.isna()] = 0

# float 값들을 int로 바꾸기
collocations = collocations.astype('int')
collocations
```

```python
# '완전'과 공기하는 횟수가 많은 것들부터 보여주기
collocations.sort_values('완전', ascending=False).head(10)
```

```python
collocations.sort_values('완전히', ascending=False).head(10)
```

```python
collocations.sort_values('아주', ascending=False).head(10)
```

## TTR: Type-token ratio

> '반짝 반짝 작은 별'

+ type: 반짝, 작은, 별 --- 3개
+ token: 반짝, 반짝, 작은, 별 --- 4개

ntype/ntoken = 3/4 = 75%

> '달아 달아 밝은 달아'

+ type: 달아, 밝은 --- 2개
+ token: 4개

ntype/ntoken = 2/4 = 50%

TTR이 클수록 다양한 단어가 사용된다.



```python
# '완전/완전히/아주'와 결합하는 단어들의 token frequency
ntokens = collocations.sum(axis=0)
ntokens
```

```python
# '완전/완전히/아주'와 결합하는 단어들의 type frequency
ntypes = (collocations > 0).sum(axis=0)
ntypes
```

```python
# TTR: Type-Token Ratio
ntypes / ntokens * 100

# '완전히'가 결합할 수 있는 단어가 '완전' 및 '아주'보다 다양하다.
```

## 완전 vs. 아주

```python
wj_aj = collocations[(collocations['완전'] > 0) & (collocations['아주'] > 0)]
del wj_aj['완전히']
wj_aj['완전/아주'] = wj_aj['완전'].div(wj_aj['아주'])
wj_aj
```

```python
wj_aj.sort_values('완전/아주').head(30)
```
