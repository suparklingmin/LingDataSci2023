```python
from collections import Counter

keywords = ('ì™„ì „', 'ì™„ì „íˆ', 'ì•„ì£¼')

# dict ì´ˆê¸°í™”
collocations_dict = {} 

# 'ì™„ì „ ã…‡ã…‡' ê¼´ì—ì„œ 'ã…‡ã…‡'ì— í•´ë‹¹í•˜ëŠ” ë¶€ë¶„ì„ ëª¨ì•„ì„œ ì¶œí˜„ íšŸìˆ˜ë¥¼ ì„¸ê¸°
for word in keywords:
    collocations_dict[word] = pd.Series(Counter(forms.str.findall(fr'\b{word} (\w+)').sum()))

collocations = pd.DataFrame(collocations_dict)
collocations
```

```python
# NaN ê°’ì„ 0ìœ¼ë¡œ ë°”ê¾¸ê¸°
collocations[collocations.isna()] = 0

# float ê°’ë“¤ì„ intë¡œ ë°”ê¾¸ê¸°
collocations = collocations.astype('int')
collocations
```

## TTR: Type-token ratio

> 'ë°˜ì§ ë°˜ì§ ì‘ì€ ë³„'

+ type: ë°˜ì§, ì‘ì€, ë³„ --- 3ê°œ
+ token: ë°˜ì§, ë°˜ì§, ì‘ì€, ë³„ --- 4ê°œ

ntype/ntoken = 3/4 = 75%

> 'ë‹¬ì•„ ë‹¬ì•„ ë°ì€ ë‹¬ì•„'

+ type: ë‹¬ì•„, ë°ì€ --- 2ê°œ
+ token: 4ê°œ

ntype/ntoken = 2/4 = 50%

TTRì´ í´ìˆ˜ë¡ ë‹¤ì–‘í•œ ë‹¨ì–´ê°€ ì‚¬ìš©ëœë‹¤.



```python
# 'ì™„ì „/ì™„ì „íˆ/ì•„ì£¼'ì™€ ê²°í•©í•˜ëŠ” ë‹¨ì–´ë“¤ì˜ token frequency
ntokens = collocations.sum(axis=0)
ntokens
```

```python
# 'ì™„ì „/ì™„ì „íˆ/ì•„ì£¼'ì™€ ê²°í•©í•˜ëŠ” ë‹¨ì–´ë“¤ì˜ type frequency
ntypes = (collocations > 0).sum(axis=0)
ntypes
```

```python
# TTR: Type-Token Ratio
ntypes / ntokens * 100

# 'ì™„ì „íˆ'ê°€ ê²°í•©í•  ìˆ˜ ìˆëŠ” ë‹¨ì–´ê°€ 'ì™„ì „' ë° 'ì•„ì£¼'ë³´ë‹¤ ë‹¤ì–‘í•˜ë‹¤.
```
