```python
import pandas as pd
```

```python
utterances = pd.read_csv('../NIKL_OM_utterances.csv.tar.gz', compression='gzip', on_bad_lines='skip')
utterances
```

```python
del utterances['../NIKL_OM_utterances.csv']

utterances.dropna(inplace=True)

utterances = utterances[['id', 'form', 'speaker_id']]
utterances
```

```python
forms = utterances['form'][utterances['form'].str.contains(r'\b(?:ì™„ì „(?:íˆ)?|ì•„ì£¼)\b', regex=True)]
forms
```

```python
from collections import Counter

keywords = ('ì™„ì „', 'ì™„ì „íˆ', 'ì•„ì£¼')
collocations_dict = {}

# 'ì™„ì „ ã…‡ã…‡' ê¼´ì—ì„œ 'ã…‡ã…‡'ì— í•´ë‹¹í•˜ëŠ” ë¶€ë¶„ì„ ëª¨ì•„ì„œ ì¶œí˜„ íšŸìˆ˜ë¥¼ ì„¸ê¸°
for word in keywords:
    collocations_dict[word] = pd.Series(Counter(forms.str.findall(fr'\b{word} (\w+)').sum()))

collocations = pd.DataFrame(collocations_dict)
collocations
```

```python
collocations[collocations.isna()] = 0
collocations = collocations.astype('int')
collocations
```

```python
collocations.sort_values('ì™„ì „', ascending=False).head(10)
```

```python
collocations.sort_values('ì™„ì „íˆ', ascending=False).head(10)
```

```python
collocations.sort_values('ì•„ì£¼', ascending=False).head(10)
```

```python
ntokens = collocations.sum(axis=0)
ntokens
```

```python
ntypes = (collocations > 0).sum(axis=0)
ntypes
```

```python
# TTR: Type-Token Ratio
ntypes / ntokens * 100
```
